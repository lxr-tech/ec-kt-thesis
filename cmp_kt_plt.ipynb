```python
import json
import seaborn
import pylab as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.sans-serif'] = ['DengXian']  
mpl.rcParams['axes.unicode_minus'] = False
```

```python
with open('./results/kt_dkt_desc.json', 'r', encoding='utf8')as fp:
    kt_dkt = json.load(fp)
with open('./results/kt_dkv_desc.json', 'r', encoding='utf8')as fp:
    kt_dkvmn = json.load(fp)
with open('./results/kt_akt_desc.json', 'r', encoding='utf8')as fp:
    kt_akt = json.load(fp)
with open('./results/kt_base_desc.json', 'r', encoding='utf8')as fp:
    kt_base = json.load(fp)
with open('./results/ec+kt+mtl_desc.json', 'r', encoding='utf8')as fp:
    kt_mtl = json.load(fp)
with open('./results/ec+kt+e2e_desc.json', 'r', encoding='utf8')as fp:
    kt_e2e = json.load(fp)
```

```python
fig = plt.figure(num=1, figsize=(20, 10), dpi=150, clear=True)
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, hspace=0.3, wspace=0.3)
ax1 = fig.add_subplot(2, 4, 1)
ax2 = fig.add_subplot(2, 4, 2)
ax3 = fig.add_subplot(2, 4, 3)
ax4 = fig.add_subplot(2, 4, 4)
ax5 = fig.add_subplot(2, 4, 5)
ax6 = fig.add_subplot(2, 4, 6)
ax7 = fig.add_subplot(2, 4, 7)
ax8 = fig.add_subplot(2, 4, 8)

ax_list = [ax1, ax2, ax3 , ax4, ax5, ax6, ax7, ax8, ]
key_list = ["train_kt_acc_every", "train_kt_acc_single", "train_kt_acc_right", "train_kt_acc_wrong", 
            "test_kt_acc_every", "test_kt_acc_single", "test_kt_acc_right", "test_kt_acc_wrong", ]
title_list = ["(a) train acc every", "(b) train acc single", "(c) train acc right", "(d) train acc wrong", 
              "(e) test acc every", "(f) test acc single", "(g) test acc right", "(h) test acc wrong", ]
lable_list = ['acc', 'acc', 'acc', 'acc', 'acc', 'acc', 'acc', 'acc', ]
loc_list = ['lower right', 'lower right', 'center right', 'center right', 
            'lower right', 'lower right', 'center right', 'center right', ]
color_list = [['sandy brown', 'tan', 'golden rod', 'orange', 'orange red', 'red'], ['sage', 'green', 'teal', 'dodger blue', 'cornflower blue', 'blue']]
limit_list = [(0, 0.9), (0.5, 1), (0, 1), (0.3, 1), (0, 0.9), (0.5, 1), (0, 1), (0.3, 1)]

for i in range(8):
    l1, = ax_list[i].plot(kt_dkt[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][0]],  ls='-.', lw='2')
    l2, = ax_list[i].plot(kt_dkvmn[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][1]], ls='-.', lw='2')
    l3, = ax_list[i].plot(kt_akt[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][2]], ls='-.', lw='2')
    l4, = ax_list[i].plot(kt_base[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][3]], lw='2')
    l5, = ax_list[i].plot(kt_mtl[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][4]], lw='2')
    l6, = ax_list[i].plot(kt_e2e[key_list[i]], c=seaborn.xkcd_rgb[color_list[(i // 4) % 2][5]], lw='2')
    ax_list[i].set_title(title_list[i], fontsize=16)
    ax_list[i].set_xlabel('iteration', fontsize=16)
    ax_list[i].set_ylabel(lable_list[i], fontsize=16)
#     ax_list[i].set_ylim(limit_list[i])
    ax_list[i].legend([l1, l2, l3, l4, l5, l6], ['DKT', 'DKVMN', 'AKT', 'KT base', 'EC+KT+mtl', 'EC+KT+e2e'], loc=loc_list[i], fontsize=10)  

plt.savefig('./results/cmp_kt/cmp_kt_plt.jpg')
```
